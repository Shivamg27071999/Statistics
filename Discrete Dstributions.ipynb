## Bernouli Distribution

- The bernouli distribution modes a random experiment based on two possible outcomes. Sometimes the two possible outcomes is knows as "Success" or "False"
    - Mathemetical representation of Bernoulli Distribution
    - X~B(1,P), here P stands for probability , X is Random variables & B is bernoulli distribution
    - 

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import math as mt
```

```python
#statistical operation library 
import scipy
from scipy import stats as st
```

```python
bern = st.bernoulli.rvs(p=0.3, size = 10000)
```

```python
bern
```

```python
#value count is not for numpy array
bern.value_counts()
```

```python
# for that we need to convert numpy as series
pd.Series(bern).value_counts()
```

```python
# for barplot
pd.Series(bern).value_counts().plot(kind="bar")
```

## calculating PMF  for 0 and 1

```python
st.bernoulli.pmf(k=0, p = 0.3) #probability of failure
```

```python
st.bernoulli.pmf(k=1, p = 0.3) #probability of success
```

## Binomial Distribution 

- The binomial distribution models the number of success (X) in a fixed number of independent trials
- The methemetical representation of Binomial Distribution
    - P(X=x)= nCx (P)*(1-P)n-x
    - here P is Probability
    - x is no. of success
    - n is no. of bernoulli trials

```python
binoms =st.binom.rvs(n=10, p=0.4,size=10000)
```

```python
pd.Series(binoms).value_counts()
```

```python
# for barplot
pd.Series(binoms).value_counts().plot(kind="bar")
```

```python
# for barplot
pd.Series(binoms).value_counts().sort_index().plot(kind="bar")
```

```python
#length of binoms
len(binoms)
```

### Poisson Distibution

- The Poisson Distibution Models the number of events in a fixed number of events or trials
- methemetical representation of a Poisson Distibution
    - X~P(Lamda)
    - PMF is P(X=x)=f(x)= (e pow(-lamda))*(Lmada pow(x))/x!

### Example Poisson Distribution
- on an average for attending class is 3, find the probability for 4 student attend the class
- here lamda is 3, and x is 4
- remember value of e is 2.71.....

```python
# value of e prove by numpy
np.e
```

```python
# solving the example with scipy
```

```python
l=3 # lamda
x=4 # poisson distribution
((np.e**(-l))*(l**x)/mt.factorial(x))
```

```python
# generate some random variable for poisson distribution
pois = st.poisson.rvs(mu=20,size = 10000)
```

```python
# converting numpy data to series
pd.Series(pois).value_counts()
```

```python
#making plot for this poisson distribution
pd.Series(pois).value_counts().sort_index().plot(kind="bar", figsize=(10,3))
```

### Example 2 on an average for selecting data scientis in a month is 40, what should be the probability for selectign 50 data sceintist in a month
- here lamda is 40 & x is 50

```python
#solution example 2 solution with methemetical representation
l=40 # lamda
x=50# poisson distribution
((np.e**(-l))*(l**x)/mt.factorial(x))
```

```python
# example 2 can be solved using stats
st.poisson.pmf(mu=40, k=50)
```
