### CDF -  Commulative Density Function
- here PDF when min = 0 and max = 4, so pdf for x=0 is 1/4, x=1 is 1/4, x=2 is 1/4, x=3 is 1/4
- cdf for X=4 is, pdf(x=0)+pdf(x=1)+pdf(x=2)+pdf(x=3)

```python
st.uniform.cdf(x=2, loc=0,scale=3)
```

### Example uniform Distribution
- The student marks of the classs unifromely distributed with minimum 10 marks and maximum 90 marks. where is the probability that student get 50 marks and find the commulative probability for the same

```python
# Solution for above example
# by cdf
st.uniform.cdf(x=50, loc=10,scale=90)
```

## Exponential Distribution
- The Exponential distribution models the time until an event occurs (events occurs continously and independently)
- if the data follows exponential distribution
    - X~Exponential(lambda) :. here lambda is constant average rate
- PDF for Exponential Distribution 
    - PDF = f(lambda) = e pow(-lamda*x); lamda>0; otherwise
