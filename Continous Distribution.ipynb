### Example -

### Normal Distribution

    - PDF = P(X=x)=f(x)=(1/(SD.Underroot.2.pie)). (e)^(-1/2(x-mean)/SD/^2)

## Empherial Rule for Normal Distribution
- if the data is following normal distribution
1. 68% = (Mean-1*SD), (Mean+1*SD)
2. 95% = (Mean-2*SD), (Mean+2*SD)
3. 99.7% = (Mean-3*SD), (Mean+3*SD)

```python
#norm.mean()-1*SD
#norm.mean()+1*SD
#norm.mean()-2*SD
#norm.mean()+2*SD
#norm.mean()-3*SD
#norm.mean()+3*SD
norms.plot(kind='density', figsize=(9,4))
plt.axvline(norms.mean(),color='red', linestyle="--", label = "mean")
plt.axvline(norms.mean()-1*norms.std(),color='purple', linestyle="--", label='M-1*SD')
plt.axvline(norms.mean()+1*norms.std(),color='purple', linestyle="--", label='M+1*SD')
plt.axvline(norms.mean()-2*norms.std(),color='green', linestyle="--", label='M-2*SD')
plt.axvline(norms.mean()+2*norms.std(),color='green', linestyle="--", label='M+2*SD')
plt.axvline(norms.mean()-3*norms.std(),color='yellow', linestyle="--", label='M-3*SD')
plt.axvline(norms.mean()+3*norms.std(),color='yellow', linestyle="--", label='M+3*SD')
plt.legend()
```

```python
# to find the values between 68% data
sum((norms>norms.mean()-1*norms.std())&(norms<norms.mean()+1*norms.std()))
```

```python
#the sum is total value and finding %
(sum((norms>norms.mean()-1*norms.std())&(norms<norms.mean()+1*norms.std()))/len(norms))*100
```

```python
# to find the values between 95% data
sum((norms>norms.mean()-2*norms.std())&(norms<norms.mean()+2*norms.std()))
```

```python
#the sum is total value and finding 95%
(sum((norms>norms.mean()-2*norms.std())&(norms<norms.mean()+2*norms.std()))/len(norms))*100
```

```python
# to find the values between 99.7% data
sum((norms>norms.mean()-3*norms.std())&(norms<norms.mean()+3*norms.std()))
```

```python
#the sum is total value and finding 95%
(sum((norms>norms.mean()-3*norms.std())&(norms<norms.mean()+3*norms.std()))/len(norms))*100
```
